```python
import numpy as np
import matplotlib.pyplot as plt

barwidth = 0.25
fig = plt.subplots(figsize = (12,8))

averaging = [94.97, 94.97, 94.97, 94.97, 94.97, 94.97]
others = [92.95, 87.92, 93.96, 92.79, 93.12, 94.79]

br1 = np.arange(len(others))
br2 = [x + barwidth for x in br1]

plt.bar(br1, others, color = 'r', width = barwidth, edgecolor = 'grey', label = 'Classic ML Algorithms')
plt.bar(br2, averaging, color = 'b', width = barwidth, edgecolor = 'grey', label = ' Ensemble Learner')
plt.xlabel('ML methods')
plt.ylabel('Accuracy')
plt.xticks([r for r in range(len(others))], ['Decision Tree', 'KNN', 'Logistic Regression', 'Naive Bayes', 'SVM', 'Random Forest'])
plt.ylim(85, 100)
plt.title('Comparison of accuracies between Averaging and other classic ML algorithms')
plt.legend()
plt.show()
```
