```python
import matplotlib.pyplot as plt

data = {'Decision Tree' : 0.9295302013422819, 'KNN' : 0.8791946308724832, 'Logistic Regression' : 0.9395973154362416, 
        'Naive Bayes' : 0.9278523489932886, 'SVM' : 0.9312080536912751, 'Random Forest' : 0.947986577181208, 
        'Max Voting' : 0.9614093959731543, 'Averaging' : 0.9496644295302014, 
        'Stacking' : 0.9513422818791947, 'Bagging' : 0.9614093959731543, 'AdaBoost' : 0.9630872483221476,
        'Gradient Boosting' : 0.9681208053691275}
methods = list(data.keys())
values = list(data.values())
fig = plt.figure(figsize = (18, 7))
 
plt.bar(methods, values, color ='green', width = 0.4)

plt.ylim(0.85, 1)
plt.xlabel("Methods used")
plt.ylabel("Accuracy")
plt.title("Machine Learning Methods and their accuracies")
plt.show()
```
